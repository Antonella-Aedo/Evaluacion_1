**4  Análisis de valores faltantes**


```python
import pandas as pd

# Creamos el DataFrame de resumen
resumen_faltantes = pd.DataFrame({
    'Columna': df.columns,
    'Total Faltantes': df.isnull().sum().values,
    'Porcentaje (%)': ((df.isnull().sum().values / len(df)) * 100).round(2)

})

# Solo mantenemos las filas donde 'Total Faltantes' es mayor a 0
resumen_faltantes = resumen_faltantes[resumen_faltantes['Total Faltantes'] > 0]

# Ordenar para que los que tienen más faltantes salgan primero
resumen_faltantes = resumen_faltantes.sort_values(by='Total Faltantes', ascending=False)

print(resumen_faltantes)
```

**Visualización de datos sin procesar**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# 1. Configuración de ruta
path_salida = '../outputs/visualizaciones/diagnostico_crudo'
os.makedirs(path_salida, exist_ok=True)

# 2. Lista de columnas numéricas
columnas_numericas = [
    'CostoAtencionCLP',
    'NumTotal', 'NumMenor1Anio', 'Num1a4Anios', 
    'Num5a14Anios', 'Num15a64Anios', 'Num65oMas'
]

for col in columnas_numericas:
    if col in df.columns:
        
        # =========================
        #  HISTOGRAMA
        # =========================
        plt.figure(figsize=(10, 6))
        
        media = df[col].mean()
        mediana = df[col].median()
        
        sns.histplot(df[col], kde=False, color="#9b59b6")
        
        # Líneas de referencia
        plt.axvline(media, color='red', linestyle='--', linewidth=2, label=f'Media: {media:.2f}')
        plt.axvline(mediana, color='green', linestyle='-', linewidth=2, label=f'Mediana: {mediana:.2f}')
        
        plt.title(f'Histograma: {col}', fontsize=14, fontweight='bold')
        plt.ylabel('Frecuencia')
        plt.xlabel(col)
        plt.legend()
        
        # Guardar histograma
        nombre_hist = f"histograma_{col}.png"
        plt.savefig(os.path.join(path_salida, nombre_hist), dpi=300, bbox_inches='tight')
        plt.show()
        
        
        # =========================
        #  BOXPLOT
        # =========================
        plt.figure(figsize=(10, 4))
        
        sns.boxplot(x=df[col], color="#e74c3c")
        
        plt.title(f'Boxplot: {col}', fontsize=14, fontweight='bold')
        plt.xlabel(col)
        
        # Guardar boxplot
        nombre_box = f"boxplot_{col}.png"
        plt.savefig(os.path.join(path_salida, nombre_box), dpi=300, bbox_inches='tight')
        plt.show()
        
        print(f"--- Gráficos individuales generados para: {col} ---")
```

Guía de Interpretación de Gráficos (Datos Crudos)

Para el diagnóstico de la base de datos urgencias_noprocesados_grupo05.csv, se han generado dos tipos de visualizaciones por cada variable numérica. A continuación, se detalla cómo interpretar cada uno y por qué presentan deficiencias visuales en esta etapa:

1. Histograma de Frecuencias (Distribución)


¿Qué busca demostrar?: Representa la "forma" de los datos. Nos permite ver dónde se concentra la mayor cantidad de registros (la montaña) y hacia dónde se desplazan los datos (sesgo).

Interpretación de la Media (Línea Roja): En un mundo ideal, la línea roja debería estar en el centro de la montaña. Si la línea está desplazada hacia un lado, significa que los valores extremos están "tironeando" del promedio, distorsionando la realidad estadística.

¿Por qué se ve de esa forma?: * Si el gráfico muestra barras muy separadas o una sola barra gigante y el resto plano, es debido a la presencia de datos ruidosos o fuera de escala.

Los espacios en blanco dentro del histograma a menudo representan rangos de valores donde no hay datos, o donde el ruido es tan alto que la escala se rompe.

2. Boxplot o Diagrama de Caja (Auditoría de Atípicos)


¿Qué busca demostrar?: Es nuestra principal herramienta de auditoría. La "caja" contiene el 50% central de los datos. Los puntos que escapan de los "bigotes" son valores atípicos (outliers).

Interpretación de Atípicos: En esta base de datos cruda, estos puntos representan posibles errores de ingreso, costos mal digitados o edades imposibles.

¿Por qué se ve de esa forma?: * En muchas columnas, el Boxplot se ve "aplastado" o extremadamente largo. Esto ocurre porque existen valores nulos o inconsistencias extremas que fuerzan a Excel/Python a usar una escala demasiado amplia.

La abundancia de puntos fuera de la caja confirma que la base de datos tiene un alto nivel de "suciedad" que impide ver la mediana con claridad.
